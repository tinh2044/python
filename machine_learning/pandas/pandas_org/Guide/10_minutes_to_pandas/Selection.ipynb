```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.arange(0,60).reshape(12,5),index=list("abcdefghijkl"), columns = list("ABCDE"))
```

#### Getting


```python
# Selecting a single column, which yields a Series, equivalent to df.A:
df[0:7]
```

```python
# Selecting via [] (__getitem__), which slices the rows:
df["a":"f"]
```

#### Selection by label

```python
# See more in Selection by Label using DataFrame.loc() or DataFrame.at().

# For getting a cross section using a label:

df.loc["a"]
```

```python
# Selecting on a multi-axis by label:
df.loc[:, ["A", "B"]]
```

```python
# Showing label slicing, both endpoints are included:
df.loc["a":"f", ["C", "D"]]
```

```python
# Reduction in the dimensions of the returned object:
df.loc["a", ["A", "E"]]
```

```python
# For getting a scalar value:
df.loc["a", "A"]
```

```python
# For getting fast access to a scalar (equivalent to the prior method):
df.at["f", "E"]
```

#### Selection by position

```python
# See more in Selection by Position using DataFrame.iloc() or DataFrame.at().

# Select via the position of the passed integers:
df.iloc[2]
```

```python
# By integer slices, acting similar to NumPy/Python:
df.iloc[0:3,0:-1]
```

```python
# By lists of integer position locations, similar to the NumPy/Python style:
df.iloc[[1,3,4], [0,3]]
```

#### Boolean indexing

```python
# Using a single column’s values to select data:
df[(df.D <10) & (df.B < 6)]
```

```python
# Using a single column’s values to select data:
df[df> 50]
```

```python
# Using the isin() method for filtering:
df[df.A.isin([0,1])]
```

#### Setting

```python
# Setting a new column automatically aligns the data by the indexes:
s1 = pd.Series([x for x in range(1, 13)], index=pd.date_range("20130102", periods=12))
s1
df["F"] = s1
df
```

```python

# Setting values by label:
df.at["a", "A"] = 45
df
```

```python
# Setting values by position:
df.iat[0, 1] = 0
df
```

```python
# Setting by assigning with a NumPy array:
df["F"] = np.array([3]*len(df))
df
```

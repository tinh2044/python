# Introduction to Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# see data set of seaborn
sns.get_dataset_names()
```

```python
# Load data set
tips_df = sns.load_dataset("tips")
tips_df
```

```python
# Histplot
sns.histplot(data=tips_df, x="tip")
```

```python
# KDE
sns.kdeplot(tips_df.total_bill)
```

```python
# Displot
sns.set_theme()
```

```python
sns.displot(data=tips_df, x="total_bill", kde=True, col='sex')
```

```python
# Bar Chart
sns.barplot(data=tips_df, x='sex', y="tip", estimator=np.mean)
```

```python
# Count Plot
sns.countplot(data=tips_df, x="day")
```

```python
# Box plot
sns.boxplot(data=tips_df, x="day", y="total_bill", hue='sex')
```

```python
# Facet Gird
tip_fg = sns.FacetGrid(data=tips_df, row="smoker", col="time")
tip_fg.map(sns.scatterplot, "total_bill", "tip")
```

```python
new_fg = sns.FacetGrid(data=tips_df, col="day", hue='smoker'
                       )
new_fg.map(sns.histplot, "total_bill")
new_fg.add_legend()
```

```python
# Joint Plot
penguins_df = sns.load_dataset("penguins")
penguins_df
```

```python
sns.jointplot(data=penguins_df, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
# Pair Plots
sns.pairplot(data=penguins_df, hue="species", )
```
